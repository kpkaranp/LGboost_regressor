```python
dataset = pd.read_csv('insurance_pre.csv')
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset)  #data pre process(converting string into number-nominal data)
dataset = dataset.astype(int)
```

```python
dataset
```

```python
no_of_rows_and_col=dataset.shape                                # to find number of rows and col's
print("Number of rows and columns:",no_of_rows_and_col)        # Here the number of rows = 1338 and the columns = 6
```

```python
dataset.columns
```

```python
indep=dataset[['age', 'sex_female','sex_male','bmi', 'charges','children','smoker_no', 'smoker_yes']]
```

```python
indep
```

```python
depend=dataset[['charges']]
```

```python
depend
```

```python
from sklearn.model_selection import train_test_split             #sklearn library used to split train set and test set 

X_train,X_test,Y_train,Y_test=train_test_split(indep, depend, test_size=0.30, random_state=0)

                                                              #Then call train_test_split function it is parameterised function
```

```python
X_train
```

```python
X_train
```

```python
import lightgbm as lgb
from lightgbm import LGBMRegressor
```

```python
regressor=LGBMRegressor()
regressor
# from sklearn.svm import SVR
# regressor= SVR(C=1.0, epsilon=0.2)
regressor.fit(X_train,Y_train)
regressor=regressor.fit(X_train,Y_train)
```

```python
y_pred=regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(Y_test,y_pred)                                         #To find the r2_score,we take actual output and predictor.
```

```python
r_score
```

```python
import pickle                             #Pickle is uesd for save the modle(pickle filename extention is 'sav')
filename="finalized_model_LGboost.sav"        # creat file    
```
